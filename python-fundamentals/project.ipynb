```python
num_1 = 10
num_2 = 10

if num_1 > num_2:
    print("El numero 1 es mayor que el numero 2")
elif num_1 < num_2:
    print("El numero 1 es menor que el numero 2")
else:
    print("Los numeros son iguales")
```

```python
#Condicionales anidados

num_1 = 20
num_2 = 10

if num_1 > num_2:
    if num_1 > 10:
        print("El numero 1 es mayor que el numero 2 y mayor que 10")
    print("El numero 1 es mayor que el numero 2")
elif num_1 < num_2:
    print("El numero 1 es menor que el numero 2")
else:
    print("Los numeros son iguales")
```

```python
#Operadores logicos

# and: se utiliza para evaluar si dos condiciones son verdaderas
# or: se utiliza para evaluar si al menos una condicion es verdadera
# not: se utiliza para evaluar si una condicion es falsa

num_1 = 20
num_2 = 20
numeros = "yes "

if num_1 and num_2 == 20:
    print(f"{numeros}El numero 1 y el numero 2 son iguales a 20")
else:
    print("Los numeros no son iguales a 20")
```

```python
#Juego piedra papel o tijeras

nombre_1 = input("Ingresa el nombre del jugador 1: ")
nombre_2 = input("Ingresa el nombre del jugador 2: ")

jugador_1 = input("Jugador 1: Elige piedra, papel o tijeras: ")
jugador_2 = input("Jugador 2: Elige piedra, papel o tijeras: ")

if jugador_1 == jugador_2:
    print("Empate")
elif (jugador_1 == "piedra" and jugador_2 == "tijeras") or \
    (jugador_1 == "papel" and jugador_2 == "piedra") or \
    (jugador_1 == "tijeras" and jugador_2 == "papel"):
    print(f"{nombre_1} gana")
elif (jugador_2 == "piedra" and jugador_1 == "tijeras") or \
    (jugador_2 == "papel" and jugador_1 == "piedra") or \
    (jugador_2 == "tijeras" and jugador_1 == "papel"):
    print(f"{nombre_2} gana")
else:
    print("Elección no válida.")
```
